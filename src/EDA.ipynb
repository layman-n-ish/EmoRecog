```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
df_old = pd.read_csv('linear_features.csv')
df = df_old.drop(['A'], axis=1)
Y = df['107']
X = df.drop(['107'], axis=1)

print(df.head())
```

```python
df.info() 
#No null values, no categorical columns 
```

```python
df.describe()
#nothing out of the blue
```

```python
df.hist(bins=20, figsize=(20, 20))
plt.show()
#the heavy-tailed distributions can be transformed using 'log' or 'sqrt' transformations
#some features have some outliers (suggested by the bins being composed in one place)
```

```python
corr_matrix = df.corr()
print(corr_matrix)
#some features have very high correlation with the other ones, this has to be avoided
```

```python
from sklearn.decomposition import IncrementalPCA
pca = IncrementalPCA(n_components=2, batch_size=3)
pca.fit(df)
pca_red = pca.transform(df)
print(pca_red)

from matplotlib import pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
plt.scatter(pca_red[:,0], pca_red[:,1], c=Y, cmap = plt.get_cmap("jet")) 
plt.show()
```
